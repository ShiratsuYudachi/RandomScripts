```python
# Q5
# 22:15

R_sets = ((2,9,10),(3,7,15),(5,12,12),(15,20,10),(19,24,8))


def isInside_pair(gamma, gamma_pair):
    # gamma_pair[0] is the left-top point of a rectangle, [1] is the right-bottom point
    # denote them as lTop and rBottom
    # if lTop.y>gamma.y and lTop.x<gamma.x<rBottom.x
    if gamma_pair[0][1]>gamma[1] and (gamma_pair[0][0]<gamma[0]<gamma_pair[1][0]):
        return True
    return False
    # Linear time

def isInside(gamma, gamma_array):
    for i in range(len(gamma_array)-1):
        if isInside_pair(gamma,[gamma_array[i],[gamma_array[i+1][0],0]]):
            return True
    return False
    # O(gamma_pair)

def solve(R_sets):
    n = len(R_sets)
    if n==1:
        # denote the only element as R
        # return list of pairs [ (R.l, R.h), (R.r, 0) ]
        return [(R_sets[0][0],R_sets[0][2]),(R_sets[0][1],0)]
    
    gamma_1 = solve(R_sets[:n//2])
    gamma_2 = solve(R_sets[n//2:])

    # merge
    for i in gamma_2: # O(n/2)
        if isInside(i,gamma_1): # O(n/2)
            if i[1]==0:
                while isInside(i,gamma_1):
                    i = (i[0],i[1]+1)
            else:
                while isInside(i,gamma_1):
                    i = (i[0]+1,i[1])
        gamma_1.append(i)
        gamma_1.sort(key=lambda x: x[0])
    return gamma_1


solve(R_sets) 
```

```python
def solve(buildings):
    points = []
    for building in buildings:
        points.append([building[0], -building[2]])
        points.append([building[1], building[2]])
    
    points.sort()
    result = []
    heights = [0]  # Initial ground height
    prev_max_height = 0
    
    for point in points:
        if point[1] < 0:  # If it's the start of a building
            # Add the height of the building
            heights.append(-point[1])
        else:  # If it's the end of a building
            # Remove the height of the building
            heights.remove(point[1])
        current_max_height = max(heights)
        
        # If the current max height is different from the previous one
        if current_max_height != prev_max_height:
            result.append([point[0], current_max_height])
            prev_max_height = current_max_height
    
    return result

    
buildings = list(map(list, R_sets))



# 调用getSkyline方法，并打印结果
skyline = solve(buildings)
print(skyline)
```

```python
def getSkyline(buildings):
    # Divide and Conquer method
    def mergeSkyline(left, right):
        # Initialize variables
        h1, h2 = 0, 0  # Heights of the current skylines
        i, j = 0, 0  # Indices for left and right skylines
        combined = []  # The combined skyline

        # Merge the two skylines together
        # IMPORTANT: merge时考虑同时遍历两个！！
        while i < len(left) and j < len(right):
            x = 0
            if left[i][0] < right[j][0]:
                x, h1 = left[i]
                i += 1
            elif left[i][0] > right[j][0]:
                x, h2 = right[j]
                j += 1
            else:  # Equal x coordinates, pick higher height
                x, h1 = left[i]
                h2 = right[j][1]
                i += 1
                j += 1
            height = max(h1, h2)
            if not combined or combined[-1][1] != height:
                combined.append([x, height])

        # Append the remaining skyline points
        combined.extend(left[i:] or right[j:])
        return combined

    def findSkyline(buildings):
        # Base case: if there is only one building, return its skyline
        if not buildings:
            return []
        if len(buildings) == 1:
            x_start, x_end, height = buildings[0]
            return [[x_start, height], [x_end, 0]]

        # Recursive case: divide and conquer
        mid = len(buildings) // 2
        leftSkyline = findSkyline(buildings[:mid])
        rightSkyline = findSkyline(buildings[mid:])
        return mergeSkyline(leftSkyline, rightSkyline)

    return findSkyline(buildings)

# Example usage
buildings = [[2,9,10], [3,7,15], [5,12,12], [15,20,10], [19,24,8]]
print(getSkyline(buildings))
```
