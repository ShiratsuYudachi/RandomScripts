```python
import random

def partition(A, p, r):
    pivot = A[r]
    i = p - 1
    for j in range(p, r):
        if A[j] <= pivot:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i+1], A[r] = A[r], A[i+1]
    return i + 1

def quick_select(A, p, r, i):
    if p == r:
        return A[p]
    # 随机选择一个枢轴并与末尾元素交换
    pivot_index = random.randint(p, r)
    A[pivot_index], A[r] = A[r], A[pivot_index]
    # 对数组进行划分
    q = partition(A, p, r)
    k = q - p + 1
    # 确定递归的方向
    if i == k:
        return A[q]
    elif i < k:
        return quick_select(A, p, q-1, i)
    else:
        return quick_select(A, q+1, r, i-k)


def my_quick_select(A, p, r, i):
    if p == r:
        return A[p]
    pivot_index = random.randint(p, r)
    A[pivot_index], A[r] = A[r], A[pivot_index]
    
    q = partition(A, p, r)
    
    if i == q+1:
        return A[q]
    elif i < q+1:
        return my_quick_select(A, p, q-1, i)
    else:
        return my_quick_select(A, q+1, r, i)


# 测试代码
A = [3, 2, 1, 5, 4]
n = len(A)
i = 2 # 查找第3小的元素
print("第", i, "小的元素是:", my_quick_select(A, 0, n-1, i))

# 测试用例
test_cases = [
    ([10, 4, 5, 8, 6, 11, 26], 3),  # 中等大小的数组，寻找第3小的元素
    ([1, 2, 3, 4, 5, 6, 7, 8, 9], 5),  # 已排序数组，寻找中位数
    ([9, 8, 7, 6, 5, 4, 3, 2, 1], 7),  # 逆序数组，寻找第7小的元素
    ([5, 3, 3, 3, 5, 5, 2], 4),  # 有重复元素的数组，寻找第4小的元素
    ([100], 1),  # 单元素数组，寻找第1小的元素
    ([4, 6, 3, 2, 5, 1, 9, 8, 7], 9),  # 中等大小的数组，寻找最大元素
]

# 使用quick_select函数进行测试
for A, i in test_cases:
    print("数组:", A)
    print(f"Example 第 {i} 小的元素是:", quick_select(A, 0, len(A)-1, i))
    print(f"My 第 {i} 小的元素是:", my_quick_select(A, 0, len(A)-1, i))
    print("-" * 30)
```
