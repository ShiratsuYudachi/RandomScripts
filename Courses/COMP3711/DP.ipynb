```python
class Solution:
    
    def isMatch(self, s: str, p: str) -> bool:
        s = s+"."
        p = p+"."
        
        ls = len(s)
        lp = len(p)
        
        iS = 0
        iP = 0
        
        connectedToStart = [[False] * len(s) for _ in range(len(p))]
        for iP in range(len(p)):
            for iS in range(len(s)):
                if iP==0 and iS==0:
                    connectedToStart[iP][iS] = p[iP]=="*" or p[iP]==s[iS] or p[iP]=="?"

                if not connectedToStart[iP][iS]:
                    continue

                if p[iP]=="*":
                    if iS+1<len(s):
                        connectedToStart[iP][iS+1] = True
                    if iP+1<len(p):
                        connectedToStart[iP+1][iS] = True # to down
                    if iP+1<len(p) and iS+1<len(s):
                        connectedToStart[iP+1][iS+1] = True
                elif p[iP]=="?":
                    if iP+1<len(p) and iS+1<len(s):
                        connectedToStart[iP+1][iS+1] = True # to down right
                else:
                    if iP+1<len(p) and iS+1<len(s):
                        connectedToStart[iP+1][iS+1] = p[iP]==s[iS]

        return connectedToStart[lp-1][ls-1]



solution = Solution()
print(solution.isMatch("adceb", "*a*b"))
```
