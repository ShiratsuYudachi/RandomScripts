```python
import matplotlib.pyplot as plt
import numpy as np

# Given coefficients for k
coefficients = {
    -3: 2 * np.exp(-1j * np.pi / 4),
    3: 2 * np.exp(1j * np.pi / 4),
    -1: 5j,
    1: -5j,
    0: 1  # DC component
}

# k values
k_values = np.arange(-5, 6, 1)
# Magnitudes of a_k for plotting
magnitudes = np.array([abs(coefficients[k]) if k in coefficients else 0 for k in k_values])
# Phases of a_k for plotting
phases = np.array([np.angle(coefficients[k]) if k in coefficients else 0 for k in k_values])

# Plot |a_k| versus k
plt.figure(figsize=(14, 4))
plt.stem(k_values, magnitudes, basefmt=" ")
plt.title('Magnitude of $a_k$ versus $k$')
plt.xlabel('$k$')
plt.ylabel('$|a_k|$')
plt.grid(True)
plt.show()

# Plot ∠a_k versus k
plt.figure(figsize=(14, 4))
plt.stem(k_values, phases, basefmt=" ")
plt.title('Phase of $a_k$ versus $k$')
plt.xlabel('$k$')
plt.ylabel('∠$a_k$ (radians)')
plt.grid(True)
plt.show()
```
